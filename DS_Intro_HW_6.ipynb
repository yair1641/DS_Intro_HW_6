```python

import requests
from bs4 import BeautifulSoup
import pandas as pd
import matplotlib.pyplot as plt
import string
import numpy as np
```

```python
data= pd.read_csv("C:\\Users\\user\\My Notebooks\\LP_destinations.csv")

data
```

```python
#Q1
data['num_of_words']= data.Description.apply(lambda x: len(str(x).split(' ')))

data
```

```python
#Q2
def remove_punctuations(text):
    for punctuation in string.punctuation:
        text = text.replace(punctuation, '')
    return text
# Apply to the DF series
data['No_punct_description'] =data['Description'].apply(remove_punctuations)
data
```

```python
#Q3
data['has_restaurants']= data.Description.str.count("restaurant")
data
```

```python
#Q4
data['has_museums']= data.Description.str.count("museum")
data
```

```python
#Q5
data["has_beaches"]=data['Description'].str.count("beach")+data['Description'].str.count("ocean")+data['Description'].str.count("sea")
data
```

```python
#Q6
avg_words=data['num_of_words'].mean()

avg_words
```

```python
#Q7
longest_description=data['num_of_words'].max()
longest_description
longest_city_description=data.loc[np.where(data['num_of_words']==longest_description),'city']
longest_city_description
```

```python
#Q8
data['num_of_words'].plot.hist(figsize=(10,8),bins=len(data))
```
